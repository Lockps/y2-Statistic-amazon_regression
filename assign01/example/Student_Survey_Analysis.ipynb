## 2. Data Preprocessing (data cleaning)

```python
# Drop columns that are entirely empty
df = df.dropna(axis=1, how='all')
df.head()
```

### 2.1 Splitting Multiple Selections into Separate Rows

This can be done using the explode() function in pandas, but before that, we need to convert the comma-separated strings to lists and strip() method removes any leading, and trailing whitespaces.

```python
# Convert the comma-separated strings to lists
df['Hobbies and Interests'] = df['Hobbies and Interests  เลือกงานอดิเรกที่ชอบทำ (เลือก 3 อย่าง)'].str.split(',')
df['Favorite Apps'] = df['Favorite Apps  Apps ที่ชื่นชอบและใช้มากสุด 3 ลำดับ'].str.split(',')

# Explode the lists into separate rows
hobbies_exploded = df.explode('Hobbies and Interests')
apps_exploded = df.explode('Favorite Apps')

hobbies_exploded['Hobbies and Interests'] = hobbies_exploded['Hobbies and Interests'].str.strip()
apps_exploded['Favorite Apps'] = apps_exploded['Favorite Apps'].str.strip()

df.head()
```

### 2.3 Outlier Handling (Age)

```python
# Identifying Outliers
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'age' is your column of interest
plt.figure(figsize=(5, 3))
plt.title('Age Distribution before filter out')
plt.xlabel('Age')
sns.boxplot(x=df['อายุ (ใส่แต่ตัวเลข)'])


```

#### Removing Outlier by statistics method (IQR)

```python
# Calculate IQR
Q1 = df['อายุ (ใส่แต่ตัวเลข)'].quantile(0.25)
Q3 = df['อายุ (ใส่แต่ตัวเลข)'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_filtered = df[(df['อายุ (ใส่แต่ตัวเลข)'] >= lower_bound) & (df['อายุ (ใส่แต่ตัวเลข)'] <= upper_bound)]

plt.figure(figsize=(5, 3))
plt.title('Age Distribution after filter out')
plt.xlabel('Age')
sns.boxplot(x=df_filtered['อายุ (ใส่แต่ตัวเลข)'])
```

## 4. Sampling Random Samples and **CLT**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# copy df_filtered to newdf for further analysis
newdf = df_filtered.copy()

# Parameters
n_samples = 93  # Total number students
sample_size = 30  # Size of each sample

# List to store the sample means
sample_means = []

# Loop to draw samples and calculate means
for _ in range(n_samples):
    sample = newdf['Social Media Usage Numeric'].sample(n=sample_size, replace=True)  # Drawing a sample
    sample_means.append(sample.mean())  # Calculating the sample mean and adding it to the list

# Calculate the overall mean and variance of the sample means
overall_mean = np.mean(sample_means)
overall_variance = np.var(sample_means, ddof=1)  # ddof=1 for sample variance

# Calculate the population mean and variance
population_mean = newdf['Social Media Usage Numeric'].mean()
population_var = newdf['Social Media Usage Numeric'].var()

print(f"Overall Mean of Sample Means: {overall_mean:.2f}")
print(f"Overall Variance of Sample Means: {overall_variance:.2f}")
print(f"Overall Mean of Population Means: {population_mean:.2f}")
print(f"Overall Variance of Population Means: {population_var:.2f}")

# Calculate the bounds for one standard deviation from the mean
one_sd_above = overall_mean + overall_variance
one_sd_below = overall_mean - overall_variance

# Calculate the Standard Error of the Mean (SEM)
s = math.sqrt(overall_variance)
SEM = s / math.sqrt(sample_size)
print(f"Standard Error of the Mean (SEM): {SEM:.2f}")


# Plotting the distribution of sample means
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, kde=True, stat='density', bins=20)
plt.xlabel('Mean Social Media Usage')
plt.ylabel('Density')
plt.title('Distribution of Sample Means vs. Population Mean')
plt.axvline(overall_mean, color='red', linestyle='--', label=f'Mean = {overall_mean:.2f}')

# Add the population mean to the histogram plot
plt.axvline(population_mean, color='g', linestyle='-', label=f'Population Mean = {population_mean:.2f}')

# Adding lines for one standard deviation above and below the mean
plt.axvline(one_sd_above, color='orange', linestyle='--', label=f'Mean + 1 SD = {one_sd_above:.2f}')
plt.axvline(one_sd_below, color='orange', linestyle='--', label=f'Mean - 1 SD = {one_sd_below:.2f}')
plt.legend()

plt.show()
```

### 4.2 Z-score (Normalization Data)-Simulate normally distributed data

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Simulate normally distributed data
np.random.seed(0)  # For reproducibility
original_data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, SD=10, n=1000

# Calculate z-scores of the simulated data
z_scores = zscore(original_data)

# Plot the original data
plt.figure(figsize=(7, 3))
plt.subplot(1, 2, 1)
sns.histplot(original_data, kde=True)
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the standardized data (z-scores)
plt.subplot(1, 2, 2)
sns.histplot(z_scores, kde=True)
plt.title('Standardized Data Distribution (Z-Scores)')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')

# Show mean and SD lines for the z-scores
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(1, color='green', linestyle='dashed', linewidth=2, label='+1 SD')
plt.axvline(-1, color='blue', linestyle='dashed', linewidth=2, label='-1 SD')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

### 4.3 Z-score (Normalization Data)-Social Media Usage

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'Social Media Usage Numeric' is the column of interest

# Calculate the mean and standard deviation of the 'Social Media Usage Numeric' column
mean_social_media_usage = newdf['Social Media Usage Numeric'].mean()
std_social_media_usage = newdf['Social Media Usage Numeric'].std()

# Calculate the z-score for each data point
newdf['Social Media Usage Z-Score'] = newdf['Social Media Usage Numeric'].apply(
    lambda x: (x - mean_social_media_usage) / std_social_media_usage
)

# Display the first few entries of the new z-score column
print(newdf[['Social Media Usage Numeric', 'Social Media Usage Z-Score']].head())

# Plot the distribution of z-scores
plt.figure(figsize=(5, 3))
sns.histplot(newdf['Social Media Usage Z-Score'], kde=True)
# Draw lines for the mean and +/- 1 standard deviation
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(1, color='green', linestyle='dashed', linewidth=2, label='+1 SD')
plt.axvline(-1, color='blue', linestyle='dashed', linewidth=2, label='-1 SD')


plt.title('Distribution of Z-Scores for Social Media Usage')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
